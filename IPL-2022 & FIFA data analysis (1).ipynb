```python
#*fifa dataset*#
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
m=pd.read_csv("fifa.csv")
m.head()
```

```python
for col in m.columns:
    print(col)
```

```python
m.shape
```

```python
m["Nationality"].value_counts()[0:10]
```

```python
plt.figure(figsize=(15,5))
plt.bar(list(m["Nationality"].value_counts()[0:5].keys()),list(m["Nationality"].value_counts()[0:5]),color="black")
plt.show()
```

```python
player_income=m[["Name","Wage"]]
player_income.head()
```

```python
player_income=player_income.sort_values(by=["Wage"],ascending=False)
player_income.head()
```

```python
plt.figure(figsize=(15,5))
plt.bar(list(player_income["Name"][0:5].value_counts().keys()),list(player_income["Name"][0:5].value_counts()),color="black")
plt.show()
```

```python
player_finishing=m[["Name","Finishing"]]
player_finishing.sort_values(by=["Finishing"],ascending=False).head()
```
